```python
#text preprocessing
def clean_text(text):
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text


def text_preprocessing(text):
    tokenizer = nltk.tokenize.RegexpTokenizer(r'\w+')
    nopunc = clean_text(text)
    tokenized_text = tokenizer.tokenize(nopunc)
    #remove_stopwords = [w for w in tokenized_text if w not in stopwords.words('english')]
    combined_text = ' '.join(tokenized_text)
    return combined_text

df['text_clean'] = df['text'].apply(str).apply(lambda x: text_preprocessing(x))

from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer

#top 10 words
def get_top_n_words(corpus, n=None):
    vec = CountVectorizer(stop_words = 'english').fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:n]
```

```python
#This is just the list of all the unigrams used, which tells us the most frequently used words and which can also be created for bigrams and so on
#For a better picture of the words used in the tweets, we can include a wordcloud for all the unique words used throughout the dataset
unigrams = get_top_n_words(df['text_clean'],10)
unigram_df = pd.DataFrame(unigrams, columns = ['Text' , 'count'])
#Plotting
plt.figure(figsize=(9,10))
sns.barplot(unigram_df["count"],unigram_df["Text"])
plt.title("Top 10 Unigrams")
plt.xlabel("Counts")
plt.ylabel("Text")
plt.show()
```

```python
#This tells us about the habits of a person, the frequency of tweets increases over the weekend and drops to at the start as per the work schedule for
#most of them
#this can be further combines with the verified users and what kind of tweet they post (positive or negative) and during what time of the week
#we will consider mostly the verified users because they have a larger reach to the population as compared to others, and the sentiment of their
#tweet will affect the readers day in a way, if it is positive the reader will have a positive outlook towards the situation
#Since the pandemic is a little tough for everyone, online positivity will help us look at the brighter side 
df['date'] = pd.to_datetime(df['date'])
df['dayofweek'] = df['date'].dt.dayofweek
df.head()
figsize = (10, 5)
daily_tweets = df.groupby(["dayofweek"])["text"].count().reset_index()
daily_tweets.columns = ['dayofweek', 'tweet_count']
sns.lineplot(x = 'dayofweek', y = 'tweet_count', data = daily_tweets)
plt.title('Tweets count by day of the week (Monday-Sunday)')
plt.show()
```

```python
#to depict the sentiment of each tweet as positive or negative
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
sid = SentimentIntensityAnalyzer()

def get_score(text):
    dict_res = sid.polarity_scores(text)
    return dict_res["compound"]

df["Score"] = df["text_clean"].apply(lambda x: get_score(x))   
sid = SentimentIntensityAnalyzer()

df['sc_ores'] = df['text'].apply(lambda review: sid.polarity_scores(review))

df.head()

df['compound']  = df['sc_ores'].apply(lambda score_dict: score_dict['compound'])

df.head()

df['comp_score'] = df['compound'].apply(lambda c: 'pos' if c >=0 else 'neg')

df.head()

sentiment_dist = df['comp_score'].value_counts()
```

```python
#this shows the sentiment of the tweet the verified user posts, which is more negative than positive
df_filter = df.loc[df["comp_score"]=="pos",:]
df_pos=df_filter.groupby('user_verified')['comp_score'].count().reset_index()

df_filter = df.loc[df["comp_score"]=="neg",:]
df_neg=df_filter.groupby('user_verified')['comp_score'].count().reset_index()

merge_df=pd.merge(df_pos,df_neg,how="left")
ax = sns.boxplot(x="user_verified", y="comp_score", data=merge_df)
```
